```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.replace(" ", pd.NA, inplace=True)
df.dropna(inplace=True)
df
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)            
df
```

# 1. Univariate Analysis

# Churn Distributuion

```python
sns.countplot(x='Churn', hue='Churn', data=df, palette='Set2', legend=False)
plt.title('Customer Churn Distribution')
plt.show()
```

# Contract Type

```python
sns.countplot(x='Contract', hue='Contract', data=df, palette='Set1', legend=False)
plt.title('Contract Type Distribution')
plt.show()
```

# 2. Bivariate Analysis

# Churn by Contract Type

```python
sns.countplot(x='Contract', hue='Churn', data=df, palette='Set3')
plt.title('Churn Rate by Contract Type')
plt.show()
```

# Churn by Internet Service

```python
sns.countplot(x='InternetService', hue='Contract', data=df, palette='cool')
plt.title('Churn by Internet Sevrice')
plt.show()
```

# Tenure Distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(df['tenure'], bins=30, kde=True)
plt.title('Customer Tenure Distribution')
plt.show()
```

# Monthly Charges Vs Churn

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df, palette='Accent')
plt.title('Monthly Charges vs Churn')
plt.show()
```

# Correlations and Insights

```python
# Convert all categorical variables for correlation
df_corr = df.copy()
for col in df_corr.select_dtypes(include='object'):
    df_corr[col] = pd.factorize(df_corr[col])[0]

plt.figure(figsize=(15, 10))
sns.heatmap(df_corr.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
